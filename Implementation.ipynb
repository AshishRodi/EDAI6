```python
import pandas as pd
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers # type: ignore
import keras_tuner as kt  # Import Keras Tuner for hyperparameter tuning

# Load the cleaned data files
data_file_path = "cleaned_processed_numeric_data.csv"
target_file_path = "cleaned_target.csv"
original_data_path = "Data.csv"

data_df = pd.read_csv(data_file_path)
target_df = pd.read_csv(target_file_path)
original_data = pd.read_csv(original_data_path)

# Extract Employee ID from the original dataset
emp_ids = original_data[['EmpID']]
data_df = pd.concat([emp_ids, data_df], axis=1)

# Split into training (80%) and testing (20%) sets
X_train, X_test, y_train, y_test = train_test_split(
    data_df.drop(columns=['EmpID']), target_df,  
    test_size=0.2, random_state=42, stratify=target_df
)

# Define a hyperparameter tuning function
def build_model(hp):
    model = keras.Sequential()
    model.add(layers.Dense(hp.Int('units_1', min_value=32, max_value=128, step=32), activation='relu', input_shape=(X_train.shape[1],)))
    model.add(layers.Dense(hp.Int('units_2', min_value=16, max_value=64, step=16), activation='relu'))
    model.add(layers.Dropout(hp.Float('dropout', min_value=0.1, max_value=0.5, step=0.1)))  # Dropout for regularization
    model.add(layers.Dense(1, activation='sigmoid'))
    
    model.compile(
        optimizer=keras.optimizers.Adam(hp.Choice('learning_rate', [0.001, 0.0005, 0.0001])),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model

# Initialize Keras Tuner
tuner = kt.Hyperband(build_model, objective='val_accuracy', max_epochs=30, factor=3, directory='tuner_dir', project_name='attrition_model')

# Perform hyperparameter search
tuner.search(X_train, y_train, epochs=30, validation_data=(X_test, y_test))

# Get the best hyperparameters
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]

print(f"Best Number of Neurons in Layer 1: {best_hps.get('units_1')}")
print(f"Best Number of Neurons in Layer 2: {best_hps.get('units_2')}")
print(f"Best Activation Function: {best_hps.get('activation')}")
print(f"Best Learning Rate: {best_hps.get('learning_rate')}")
print(f"Best Batch Size: {best_hps.get('batch_size')}")

# Build the model with the best hyperparameters
model = build_model(best_hps)

# Train the optimized model
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

# Evaluate the model
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Optimized Test Accuracy: {test_accuracy:.4f}")

# Make predictions
predictions = model.predict(X_test)
predictions = (predictions > 0.5).astype(int)  # Convert probabilities to binary values

# Combine predictions with Employee ID
results_df = pd.DataFrame({'EmpID': emp_ids.iloc[X_test.index]['EmpID'].values, 'Predicted Attrition': predictions.flatten()})

# Function to predict attrition for a given EmpID and identify the most significant reason
def predict_employee_attrition(emp_id):
    emp_row = results_df[results_df['EmpID'] == emp_id]
    if emp_row.empty:
        print("Employee ID not found in test data.")
    else:
        prediction = "Likely to Leave" if emp_row['Predicted Attrition'].values[0] == 1 else "Likely to Stay"
        print(f"Employee {emp_id}: {prediction}")
        
        # Identify the most significant factor if the employee is likely to leave
        if prediction == "Likely to Leave":
            emp_features = X_test.iloc[emp_row.index]
            most_significant_feature = emp_features.abs().idxmax(axis=1).values[0]  # Feature with highest impact
            print(f"Most significant reason: {most_significant_feature}")

# Function to list all employees likely to leave
def list_employees_leaving():
    leaving_employees = results_df[results_df['Predicted Attrition'] == 1]
    if leaving_employees.empty:
        print("No employees predicted to leave.")
    else:
        print("Employees predicted to leave:")
        for _, row in leaving_employees.iterrows():
            emp_id = row['EmpID']
            emp_features = X_test.iloc[results_df[results_df['EmpID'] == emp_id].index]
            most_significant_feature = emp_features.abs().idxmax(axis=1).values[0]
            print(f"Employee {emp_id}: Likely to Leave, Most Significant Reason: {most_significant_feature}")

# Example usage
emp_id_to_check = input("Enter Employee ID: ")
predict_employee_attrition(emp_id_to_check)

# Get all employees likely to leave
list_employees_leaving()
```
